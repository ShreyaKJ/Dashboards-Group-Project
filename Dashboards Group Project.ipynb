```python
Emp_2016 = pd.ExcelFile('state_M2016_dl.xlsx')
Emp_2016 = Emp_2016.parse("state_dl")
Emp_2016.columns = ['AREA',	'ST',	'STATE',	'OCC_CODE',	'OCC_TITLE',	'GROUP',	'TOT_EMP',	'EMP_PRSE',	'JOBS_1000',	'LOC QUOTIENT',	'H_MEAN',	'A_MEAN',	'MEAN_PRSE',	'H_PCT10',	'H_PCT25',	'H_MEDIAN',	'H_PCT75',	'H_PCT90',	'A_PCT10',	'A_PCT25',	'A_MEDIAN',	'A_PCT75',	'A_PCT90',	'ANNUAL',	'HOURLY']
Emp_2016 = Emp_2016[['STATE', 'OCC_CODE', 'OCC_TITLE', 'GROUP', 'TOT_EMP', 'EMP_PRSE', 'JOBS_1000', 'LOC QUOTIENT', 'H_MEAN', 'A_MEAN']]
Emp_2016['YEAR'] = 2016
Emp_2016.head()
```

Merging all files of to bring 6 years data together.
Note: These files have data in 3 hierarchies, which can be differentiated by the columns - 'GROUP'. 
1. 'total' means total employed popualtion of the state
2. 'major' means total employed popualtion of the state in an occupation group
3. 'detailed' means total employed popualtion of the state in an occupation title (which will be under an occupation group)

```python
Emp_File = pd.concat([Emp_2010, Emp_2011, Emp_2012, Emp_2013, Emp_2014, Emp_2015, Emp_2016])
Emp_File.head()
```

```python
Emp_File_Final = Emp_File[Emp_File['GROUP'] == 'detailed']
Emp_File_Final.head()
```

```python
Emp_File_Final['OCC_CODE']=Emp_File_Final['OCC_CODE'].apply(lambda x: pd.Series(x.split('-')))
Emp_File_Final.head()
```

```python
Group = Emp_File[Emp_File['GROUP'] == 'major']
Group = Group[Group['YEAR'] == 2016]
Group['OCC_CODE']=Group['OCC_CODE'].apply(lambda x: pd.Series(x.split('-')))
Group = Group[['STATE', 'OCC_CODE', 'OCC_TITLE']]
Group.head()
```

```python
Emp_Final = Emp_File_Final.merge(Group, on=['STATE', 'OCC_CODE'], how='left')
Emp_Final.columns = ['STATE', 'OCC_CODE', 'OCC_TITLE', 'GROUP', 'TOT_EMP', 'EMP_PRSE', 'JOBS_1000', 'LOC_QUOTIENT', 'H_MEAN', 'A_MEAN', 'YEAR', 'OCC_GRP']
Emp_Final.head()
```

Converting it to a CSV file to use as a source for Tableau for trend graphs

```python
Emp_File_Final = Emp_Final.drop('OCC_CODE',1)
Emp_File_Final = Emp_File_Final.drop('GROUP',1)
Emp_File_Final.to_csv('Emp_File_Final.csv')
Emp_File_Final.head()
```

# Wrangling for 2016 employment data only

File for 2016 with ONLY total Employment per state (No groups and no titles)


```python
Emp_Total_State = Emp_File[Emp_File['GROUP'] == 'total']
Emp_Total_State = Emp_Total_State[Emp_Total_State['YEAR'] == 2016]
Emp_Total_State = Emp_Total_State[['STATE', 'TOT_EMP']]
Emp_Total_State.head()
```

File for 2016 with ONLY employments per title per state (No groups, No Totals)

```python
Emp_Title = Emp_File[Emp_File['GROUP'] == 'detailed']
Emp_Title = Emp_Title[Emp_Title['YEAR'] == 2016]
Emp_Title = Emp_Title[['STATE', 'OCC_CODE', 'OCC_TITLE', 'GROUP', 'TOT_EMP', 'EMP_PRSE', 'JOBS_1000', 'LOC QUOTIENT', 'H_MEAN', 'A_MEAN']]
Emp_Title.head()
```

Merging title file to total file to get total employment number for each state.

```python
Emp_Title_Total_File = Emp_Title.merge(Emp_Total_State, on='STATE', how='outer')
Emp_Title_Total_File.columns = ['STATE', 'OCC_TITLE_CODE', 'OCC_TITLE', 'GROUP', 'TITLE_EMP', 'EMP_PRSE', 'JOBS_1000', 'LOC QUOTIENT', 'H_MEAN', 'A_MEAN', 'STATE_EMP']
Emp_Title_Total_File.head()
```

Adding new column - Occupation group code, to merge the total occupation employment number also.

```python
Emp_Title_Total_File['OCC_CODE'] = 0
Emp_Title_Total_File['OCC_CODE']=Emp_Title_Total_File['OCC_TITLE_CODE'].apply(lambda x: pd.Series(x.split('-')))
Emp_Title_Total_File.head()
```

File for 2016 with ONLY employments per group per state (No Titles, No Totals)

```python
Emp_Group_State = Emp_File[Emp_File['GROUP'] == 'major']
Emp_Group_State = Emp_Group_State[Emp_Group_State['YEAR'] == 2016]
Emp_Group_State['OCC_CODE']=Emp_Group_State['OCC_CODE'].apply(lambda x: pd.Series(x.split('-')))
Emp_Group_State = Emp_Group_State[['STATE',  'OCC_CODE', 'OCC_TITLE', 'TOT_EMP']]
Emp_Group_State.head()
```

Merging Title Total file to Group State file to get total employment number for each occupation group in each state

```python
Emp_Final_File = Emp_Title_Total_File.merge(Emp_Group_State, on=['STATE','OCC_CODE'], how='outer')
Emp_Final_File.head()
```

Renaming and rearranging columns

```python
Emp_Final_File.columns = ['STATE', 'OCC_TITLE_CODE', 'OCC_TITLE', 'GROUP', 'TITLE_EMP', 'EMP_PRSE', 'JOBS_1000', 'LOC_QUOTNT', 'H_MEAN', 'A_MEAN', 'STATE_EMP', 'OCC_GROUP_CODE', 'OCC_GROUP', 'GROUP_STATE_EMP']
Emp_Final_File = Emp_Final_File[['STATE', 'OCC_GROUP', 'OCC_TITLE', 'STATE_EMP', 'GROUP_STATE_EMP', 'TITLE_EMP', 'JOBS_1000', 'LOC_QUOTNT', 'H_MEAN', 'A_MEAN']]
Emp_Final_File.head()
```
